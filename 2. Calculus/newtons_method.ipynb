```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-2, 2, 100)
y = -0.5 * 9.8 * x ** 2 + 2 * x + 1
```

```python
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.grid()
plt.plot(x, y)
```

```python
def func(x):
  return -0.5 * 9.8 * x ** 2 + 2 * x + 1
```

```python
def derivative(x):
  return -0.5 * 9.8 * 2 * x + 2
```

```python
initial_x = -1

while True:
  new_x = initial_x - func(initial_x) / derivative(initial_x)
  print(new_x)

  if np.abs(new_x - initial_x) < 1e-10:
    break

  initial_x = new_x

print(f"Answer: {new_x}")
answer1 = new_x
```

```python
initial_x = 1
while True:
  new_x = initial_x - func(initial_x) / derivative(initial_x)
  print(new_x)

  if np.abs(new_x - initial_x) < 1e-10:
    break

  initial_x = new_x

print(f"Answer: {new_x}")
answer2 = new_x
```

```python
# confirming newtons method to the quadratic equation

a = -0.5 * 9.8
b = 2
c = 1

discriminant = b * b - 4 * a * c
true_answer1 = (-b + np.sqrt(discriminant)) / (2 * a)
true_answer2 = (-b - np.sqrt(discriminant)) / (2 * a)
print("newton 1:", answer1)
print("true 1:", true_answer1)
print("newton 2:", answer2)
print("true 2:", true_answer2)
```
