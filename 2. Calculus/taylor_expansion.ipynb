```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-2, 2, 100)
print(x)
```

```python
y = np.exp(x)
print(y)
```

```python
def exp_approx(x, N):
  y_all = np.zeros((N + 1, len(x)))
  print(y_all)
  for n in range(N + 1):
    y_all[n] = x**n / np.math.factorial(n)
  return y_all.sum(axis=0)

y2 = exp_approx(x, 5)
print(y2)
```

```python
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.grid()
plt.plot(x, y, label='true');
plt.plot(x, y2, label="approx")
plt.legend()
```

```python
x = np.linspace(-np.pi, np.pi, 100)
y = np.sin(x)

def sin_approx(x, N):
  y_all = np.zeros((N + 1, len(x)))
  for n in range(N + 1):
    if n % 2 == 0:
      coef = 1
    else:
      coef = -1
    nth_term = coef * (x ** (2 * n + 1)) / np.math.factorial(2 * n + 1)
    y_all[n] = nth_term
  return y_all.sum(axis=0)

y2 = sin_approx(x, 3)
```

```python
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.grid()
plt.plot(x, y, label='true');
plt.plot(x, y2, label="approx")
plt.legend()
```
