```python
train_images = np.zeros(2,)
test_images = np.zeros(2,)
```

```python
for (dirpath, dirnames, filenames) in os.walk('./dogs-vs-cats'):
  dirpath = dirpath.replace("\\","/")

  if (len(filenames) > 0):
    for i in range(len(filenames)):
      image = Image.open(f"{dirpath}/{filenames[i]}")
      np_arr = np.array(image)
      gray_image = np_arr.mean(axis=2)
      if 'train' in dirpath:
        train_images[i] = gray_image
      elif 'test' in dirpath:
        test_images[i] = gray_image
```
