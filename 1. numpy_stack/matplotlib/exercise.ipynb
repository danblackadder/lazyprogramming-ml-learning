```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1 = np.random.uniform(low=-1, high=1, size=2000)
x2 = np.random.uniform(low=-1, high=1, size=2000)
y = np.zeros(2000)

for e in range(len(y)):
  y[e] = (x1[e] <=0 and x2[e] > 0) or (x1[e] > 0 and x2[e] <= 0)
```

```python
plt.scatter(x1, x2, c=y);
```
